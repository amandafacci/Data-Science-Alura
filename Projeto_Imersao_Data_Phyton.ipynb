```python
import plotly.express as px
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df_limpo = df.dropna()
df_limpo.isnull().sum()
```

```python
df_limpo.head(10)
```

```python
df_data_scientist = df_limpo[df_limpo['job_title'] == 'Data Scientist']
df_data_sal_pais = df_data_scientist.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(df_data_sal_pais,
             x='employee_residence',
             y='salary_in_usd',
             title='Salario Anual por País - Data Scientist',
             labels={'salary_in_usd': 'Salario em USD', 'employee_residence': 'País'},
             template='plotly_white')
fig.show()
```

```python
df_data_scientist = df_limpo[df_limpo['job_title'] == 'Data Scientist']
df_data_sal_pais = df_data_scientist.groupby('employee_residence')['salary_in_usd'].mean().reset_index()
#.sort_values(): É uma função do Pandas que permite reordenar as linhas do DataFrame com base nos valores de uma ou mais colunas.
df_data_sal_pais = df_data_sal_pais.sort_values(by='employee_residence', ascending=True)

fig = px.bar(df_data_sal_pais,
             x='employee_residence',
             y='salary_in_usd',
             title='Salario Anual por País - Data Scientist',
             labels={'salary_in_usd': 'Salario em USD', 'employee_residence': 'País'},
             template='plotly_white')
fig.show()
```
